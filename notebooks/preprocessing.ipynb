### 1- transform train images into: 
- `train`/`val` (80%, 20%)
- `batch_size` 32
- `img_size` hight=100, width=50
- `class_mode` classification

```python
# imports
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# instatiate img -> data gen
train_datagen = ImageDataGenerator(validation_split=0.2, rescale=1/255)


# train generator
train_generator = train_datagen.flow_from_directory(
        '../theposturewizard/data/train',
        target_size=(100, 50),
        batch_size=32,
        subset='training',
        class_mode='binary')


# validation generator
validation_generator = train_datagen.flow_from_directory(
        '../theposturewizard/data/train',
        target_size=(100, 50),
        batch_size=32,
        subset='validation',
        class_mode='binary')

# returns 
# DirectoryIterator 's yielding tuples of (x, y) 
#    where x is a numpy array containing a batch of images with shape (batch_size, *target_size, channels) 
#    and y is a numpy array of corresponding labels.
```

### 2- Printing batches from `train_generator`

```python
import matplotlib.pyplot as plt
%matplotlib inline

def show_batch(batch_number):
    # get batch with batch_number
    batch = train_generator.__getitem__(batch_number)
    
    # labels
    labels = ['Good', 'bad']
    label_index = batch[1]
    label_index = label_index.astype('int')
    
    #preprocessed images
    imgs = batch[0]
    
    # show labeled images
    for index, img in enumerate(imgs):
    
        label = labels[label_index[index]]

        plt.figure(figsize=(5,5))
        plt.imshow(img)
        plt.title(label)
        plt.axis('off')
        plt.show()
```

```python
show_batch(3)
```
